```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
```

```python
file_url_1 = r'G:\\datamining\\data\\Melbourne Airbnb Open Data\\Melbourne Airbnb Open Data\\cleansed_listings_dec18.csv'

# 加载数据集
raw_dataframe = pd.read_csv(file_url_1)


# 查看数据数量
print(type(raw_dataframe))
print(raw_dataframe.shape[0])
print(raw_dataframe.shape[1])
feature_name = [column for column in raw_dataframe] 
print(feature_name)
# print(raw_dataframe)
raw_dataframe[:10]
```

```python
# 标称属性
for i in feature_name[1:]:
    print(i + ':')
    print(raw_dataframe[i].value_counts())
    print('\n**********************************************\n')
```

```python
#五数概括 Minimum（最小值）、Q1、Median（中位数、）、Q3、Maximum（最大值）
for i in ['beds', 'price',]:
    print('缺失值： ' + str(raw_dataframe[i].isna().sum()))
    x = raw_dataframe[i].dropna()
    Minimum=min(x)
    print(i + " 的最小值为："+ str(Minimum))
    Maximum=max(x)
    print(i + " 的最大值为："+ str(Maximum))
    Q1=np.percentile(x,25)
    print(i + " 的Q1值为："+ str(Q1))
    Median=np.median(x)
    print(i + " 的中位值为："+ str(Median))
    Q3=np.percentile(x,75)
    print(i + " 的Q3值为："+ str(Q3))
    print('\n**********************************************\n')
```

```python
plt.figure(figsize = (20, 10))
plt.subplot(1, 2, 1)
plt.hist(raw_dataframe['beds'].dropna().values.tolist())
plt.title('beds')
plt.subplot(1, 2, 2)
plt.hist(raw_dataframe['price'].dropna().values.tolist())
plt.title('price')
plt.show()
```

```python
plt.figure(figsize = (20, 10))
# 'bedrooms', 'beds', 'price'
plt.subplot(1, 2, 1)
plt.boxplot(raw_dataframe['beds'].dropna().values.tolist())
plt.title('beds')
plt.subplot(1, 2, 2)
plt.boxplot(raw_dataframe['price'].dropna().values.tolist())
plt.title('price')
plt.show()
```

```python
# 将缺失部分剔除
dataframe_1 = raw_dataframe.dropna(subset=['beds'])
print(dataframe_1.shape[0])
print(dataframe_1.shape[1])
```

```python
# 用最高频率值来填补缺失值
dataframe_2 = raw_dataframe.fillna({"beds":raw_dataframe.beds.mode(),"price":raw_dataframe.price.mode(),})
print(dataframe_2.shape[0])
print(dataframe_2.shape[1])
```

```python
# 通过属性的相关关系来填补缺失值

# 针对数值数据price和point，利用先线性模型进行预测
# 训练线性回归模型进行填补
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

train = raw_dataframe.copy().loc[~raw_dataframe['beds'].isna(),['price','beds']].reset_index(drop=True)#波浪号取反
x = train[['price']]
y = train[['beds']]
model = LinearRegression()
model.fit(x,y)

# 查看模型效果
print(f'R squared is: {r2_score(y, x)}')
print('intercept: ' + str(model.intercept_))
print('coef:      ' + str(model.coef_))

# 模型填充
dataframe_3 = raw_dataframe.copy()
dataframe_3.loc[dataframe_3['beds'].isna(),['beds']] = model.predict(dataframe_3.loc[dataframe_3['beds'].isna(),['price']])
```

```python
# 通过数据对象之间的相似性来填补缺失值

from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors = 1000, weights = "uniform")

train_knn = raw_dataframe.copy().loc[~raw_dataframe['beds'].isna(),['price','beds']].reset_index(drop=True)
x_knn = train_knn[['price']]
y_knn = train_knn[['beds']]
model_knn.fit(x_knn, np.ravel(y_knn))

# 模型填充
dataframe_4 = raw_dataframe.copy()
dataframe_4.loc[dataframe_4['beds'].isna(),['beds']] = model_knn.predict(dataframe_4.loc[dataframe_4['beds'].isna(),['price']])
```

```python
plt.figure(figsize = (20, 10))
plt.subplot(1, 5, 1)
plt.boxplot(raw_dataframe['beds'].dropna().values.tolist())
plt.title('beds (original)')
plt.subplot(1, 5, 2)
plt.boxplot(dataframe_1['beds'].dropna().values.tolist())
plt.title('beds (drop)')
plt.subplot(1, 5, 3)
plt.boxplot(dataframe_2['beds'].dropna().values.tolist())
plt.title('beds (mode)')
plt.subplot(1, 5, 4)
plt.boxplot(dataframe_3['beds'].values.tolist())
plt.title('beds (relative)')
plt.subplot(1, 5, 5)
plt.boxplot(dataframe_4['beds'].values.tolist())
plt.title('beds (similarity)')
plt.show()
```
